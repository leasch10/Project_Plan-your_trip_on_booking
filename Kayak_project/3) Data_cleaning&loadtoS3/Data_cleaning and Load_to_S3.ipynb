```python
import pandas as pd 
```

```python
df_hotel = pd.read_csv("best_hotel.csv")
df_weather = pd.read_csv("best_weather.csv")
```

```python
df_weather = pd.read_csv("best_weather.csv")
```

```python
df_weather.isnull()
```

```python
df_weather.head()
```

```python
df_weather = df_weather.drop(df_weather.columns[[0]], axis = 1)
```

```python
df_weather
```

```python
#Comme je n'ai pas utilisé le paramètre suffixe lors du merge, je dois cleaner à la main
df_weather = df_weather.rename(columns=
                        {'temp_day_x':'temp_day_max',
                         'temp_min_x':'temp_min_max',
                         'temp_max_x':'temp_max_max',
                         'feels_like_day_x':'feels_like_day_max',
                         'feels_like_morn_x':'feels_like_morn_max',
                         'sunrise_x':'sunrise_max',
                         'humidity_x':"humidity_max",
                         'wind_speed_x':'wind_speed_max',
                         'clouds_x':'clouds_max',
                         'uvi_x':'uvi_max',
                         'temp_day_y':'temp_day_mean',
                         'temp_min_y':'temp_min_mean',
                         'temp_max_y':'temp_max_mean',
                         'feels_like_day_y':'feels_like_day_mean',
                         'feels_like_morn_y':'feels_like_morn_mean',
                         'sunrise_y':'sunrise_mean',
                         'humidity_y':"humidity_mean",
                         'wind_speed_y':'wind_speed_mean',
                         'clouds_y':'clouds_mean',
                         'uvi_y':'uvi_mean'})
```

```python
df_weather.columns
```

```python

import datetime 
df_weather['sunrise_max'] = df_weather['sunrise_max'].apply(lambda x: datetime.datetime.fromtimestamp(x))
df_weather['sunrise_mean'] = df_weather['sunrise_max'].apply(lambda x: datetime.datetime.fromtimestamp(x))
```

```python
display(df_weather)
```

```python
df_cities_id = df_weather[["index", "name"]]
```

```python
df_cities_id.head()
```

```python
df_cities_id.loc[df_cities_id['name'] == 'Le-Mont-Saint-Michel', 'name'] = 'Le Mont-Saint-Michel'
```

```python
df_hotel_with_id = df_hotel.merge(df_cities_id, left_on="city_final", right_on="name")
```

```python
df_hotel_with_id.head()
```

```python
df_hotel_with_id = df_hotel_with_id.drop("name_y" ,axis = 1)
```

```python
len(df_hotel_with_id)
```

```python
df_hotel_with_id.to_csv("cleaned_hotel.csv")
```
