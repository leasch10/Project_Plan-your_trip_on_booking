```python
#Requête pour récupérer les coordonnées en fonction des villes 
list_request = []
for i in cities_list:
    payload = {
    "city": i,
    "country":"france",
    "format": "json"} 
    r = requests.get("https://nominatim.openstreetmap.org/search", params=payload)
    list_request.append(r.json())
```

```python
#récupération de la météo sur 7 jours 
weather_list = []
for index, row in df_cities.iterrows():
    try : 
        r = requests.get(f"https://api.openweathermap.org/data/3.0/onecall?lat={row['latitude']}&lon={row['longitude']}&exclude=current,minutely,hourly,alerts&appid=d4bf417204de8563a56b016c5d24f939&units=metric&lang=fr")
        weather_list.append(r.json())
    except:
        print("hello")

display(weather_list)
```

```python
# Récupération du premier élément de daily
daily_data = weather_list[0]["daily"][0]

# Récupération des clés du dictionnaire
daily_keys = list(daily_data.keys())

# Affichage des clés
print(daily_keys)
```

```python
daily_data = weather_list[0]["daily"][0]["weather"]
print(daily_data)
```

```python
#Extraire la donnée pour chaque lieu dont les données imbriquées pour en faire un dataframe
# Liste pour stocker les données
data = []

# Extraction des données pour chaque lieu
for location in weather_list:

    # Parcours de chaque jour dans 'daily'
    for day_data in location['daily']:
        # Créer un dictionnaire vide pour stocker les valeurs à extraire
        extracted_data = {}
        
        # Ajouter la latitude et longitude
        extracted_data['Latitude'] = location['lat']
        extracted_data['Longitude'] = location['lon']
        
        # Extraire les données en fonction de la liste de clés
        for key in daily_keys:
            if key in day_data:
                # Si la clé est dans le dictionnaire principal
                extracted_data[key] = day_data[key]
            elif key == 'rain':
                # Gérer le cas où 'rain' pourrait ne pas exister
                extracted_data['rain'] = day_data.get('rain', 0)
        
        for temp_key, temp_value in day_data['temp'].items():
            extracted_data[f'temp_{temp_key}'] = temp_value
        
        # Extraire les sous-dictionnaires 'feels_like' et ajouter les données sous un nom de colonne spécifique
        for feels_key, feels_value in day_data['feels_like'].items():
            extracted_data[f'feels_like_{feels_key}'] = feels_value
        # Ajouter cet ensemble de données à la liste principale
        data.append(extracted_data)

# Créer un DataFrame avec les données extraites
df = pd.DataFrame(data)

display(df)
        
```

```python
#récupération des données de weather 
df['weather_id'] = df['weather'].apply(lambda x: x[0]['id'])
df['weather_main'] = df['weather'].apply(lambda x: x[0]['main'])
df['weather_description'] = df['weather'].apply(lambda x: x[0]['description'])
df['weather_icon'] = df['weather'].apply(lambda x: x[0]['icon'])
display(df)
```
